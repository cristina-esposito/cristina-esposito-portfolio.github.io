```python
# [Step 0] import module - pandas, requests, bs4-BeautifulSoup

import pandas as pd
import requests
from bs4 import BeautifulSoup

data_dir = "../data/"
```

```python
#----------------------------------------------------------------------------
# [Step 1] Send Request

review_text=[]
user_name=[]
date=[]


for pagenum in range(1,544):
    # 1) url
    url="https://forums.edmunds.com/discussion/7526/general/x/midsize-sedans-2-0/p{0}".format(pagenum)
    
    # 2) requests.get()
    page=requests.get(url)
    page_text=page.text
    
    
#----------------------------------------------------------------------------
# [Step 2] Parsing Data
    soup=BeautifulSoup(page_text, 'html.parser')

     
    # 1) read reviewer name 
    #----------------------------------------------------
    user_name_elem=soup.find_all(class_="Username js-userCard")
    for item in user_name_elem:
        user_name.append(item.text)
    
    # 2) read time date
    #----------------------------------------------------   
    date_elem=soup.find_all(class_="MItem DateCreated")
    for item in date_elem:
        date.append(item.find('time').get("datetime"))
    
    # 3) read review text
    #----------------------------------------------------
    review_text_elem= soup.find_all(class_="Item-Body")
    
    for item in review_text_elem:
        #review_text.append(item.text)
        review_text.append(item.text.lstrip())
    
    
```

```python
#----------------------------------------------------------------------------
# [Step 3] create dataframe

final_dict={
            'User_Name': user_name,
            'Date': date, 
            'Review_Text': review_text            
            }

df=pd.DataFrame(final_dict)

df
```

```python
# save df to csv file
#----------------------------------------------------
df.to_csv(data_dir+'CARS_ALL_COMMENTS.csv', index=False)
```
